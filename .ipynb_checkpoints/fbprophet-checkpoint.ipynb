```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os,sys
from fbprophet import Prophet
```

```python
fdir = os.path.join(sys.path[0],"data/")
flist = os.listdir(fdir)
DataSet = {}
for file in flist:
    # print(file)
    Coinname = file.replace("coin_", "", 1).replace(".csv","",1)
    # print(Coinname)
    DataSet[Coinname] = pd.read_csv("data/"+file)
```

```python
# Show all the available cryptocurrencies
for key in DataSet:
    print(key)
```

```python
coin = "Litecoin"
dataframe = DataSet[coin]
```

```python
dataframe.head()
```

```python
cols = ['Date','Close']
df = dataframe[cols]
pd.options.mode.chained_assignment = None
df.rename(columns={"Date": "ds", "Close": "y"},inplace=True)
df.head()
```

```python
pd.plotting.register_matplotlib_converters()
plt.figure(figsize=(15,6))
df.set_index('ds').y.plot().figure
plt.xlabel('Date')
plt.ylabel('Price in USD')
```

```python
m = Prophet(daily_seasonality=True,seasonality_mode='multiplicative')
m.add_seasonality(name='monthly', period=30.5, fourier_order=5, prior_scale=0.02)
m.fit(df)
```

```python
future = m.make_future_dataframe(periods=365)
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].head()
```

```python
fig1 = m.plot(forecast)
forecast.plot(x='ds',y='yhat',legend=True,label='predictions',figsize=(12,6))
```

```python
fig2 = m.plot_components(forecast)
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(df['y'], forecast[:2862]['yhat']))
print("RMSE: %f" % (rmse))
```

```python
from fbprophet.diagnostics import cross_validation
df_cv = cross_validation(m, initial='730 days', period='180 days', horizon = '365 days')
```

```python
df_cv.head()
```

```python
from fbprophet.diagnostics import performance_metrics
df_p = performance_metrics(df_cv)
df_p.head()
```

```python
from fbprophet.plot import plot_cross_validation_metric
fig = plot_cross_validation_metric(df_cv, metric='rmse')
```
